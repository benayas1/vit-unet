```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
from benatools.ct import ct
from glob import glob
import os
import pydicom
import cv2
import pandas as pd
```

```python
def show_pairs(case):
    clean = sorted(glob(f'ssid/clean/{str(case).zfill(4)}-*.png'))
    noisy = sorted(glob(f'ssid/noisy/{str(case).zfill(4)}-*.png'))

    rows = 4
    columns = 2
    fig, ax = plt.subplots(rows, columns, figsize=(12,12))

    for i in range(rows):
        img_clean = cv2.imread(clean[i])
        img_noisy = cv2.imread(noisy[i])

        ax[i,0].imshow(img_clean)
        ax[i,1].imshow(img_noisy)

    plt.show()
```

```python
show_pairs(10)
```
